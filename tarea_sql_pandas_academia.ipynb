## 3. Cargar tablas a SQLite

### Punto 6. JOIN entre tablas

Muestra:
- nombre del estudiante
- nombre del curso
- estado de la inscripción

Usa `estudiantes`, `inscripciones` y `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 6 - SQL
query_6 = """
SELECT
    e.nombre AS nombre_estudiante,
    c.nombre_curso AS nombre_curso,
    i.estado
FROM inscripciones i
INNER JOIN estudiantes e ON i.estudiante_id = e.estudiante_id
INNER JOIN cursos c ON i.curso_id = c.curso_id
"""

sql_6 = run_sql(query_6)
sql_6
```

```python
# Punto 6 - pand# Unimos las 3 tablas usando las columnas que ya comprobamos que funcionan
p_6 = \
    inscripciones.merge(estudiantes, on="estudiante_id") \
    .merge(cursos, on="curso_id") \
    [["nombre", "nombre_curso", "estado"]] \
    .rename(columns={
        "nombre": "nombre_estudiante",
        "estado": "estado_inscripcion"
    })

p_6
```

### Punto 7. LEFT JOIN

Trae todos los estudiantes, incluso si no tuvieran inscripciones, mostrando:
- `estudiante_id`
- `nombre`
- `inscripcion_id`

**Entregable:**
- SQL
- pandas

> Pista: en este dataset puede que todos tengan al menos una inscripción, pero igual debes practicar el `LEFT JOIN`.

```python
# Punto 7 - SQL
query_7 = """
SELECT
    e.estudiante_id,
    e.nombre,
    i.inscripcion_id
FROM estudiantes e
LEFT JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
"""

sql_7 = run_sql(query_7)
sql_7
```

```python
# Punto 7 - pandas
p_7 = estudiantes.merge(inscripciones, on="estudiante_id", how="left") \
    [["estudiante_id", "nombre", "inscripcion_id"]]

p_7
```

### Punto 8. HAVING

Calcula cuántas inscripciones tiene cada estudiante y muestra solo aquellos con **2 o más**.

Debes mostrar:
- `estudiante_id`
- `nombre`
- `n_inscripciones`

**Entregable:**
- SQL
- pandas

```python
# Punto 8 - SQL
query_8 = """
SELECT
    e.estudiante_id,
    e.nombre,
    COUNT(i.inscripcion_id) AS n_inscripciones
FROM estudiantes e
INNER JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
GROUP BY e.estudiante_id, e.nombre
HAVING n_inscripciones >= 2
"""

sql_8 = run_sql(query_8)
sql_8
```

```python
# Punto 8 - pandas
p_8 = inscripciones.merge(estudiantes, on="estudiante_id") \
    .groupby(["estudiante_id", "nombre"])["inscripcion_id"] \
    .count() \
    .reset_index() \
    .rename(columns={"inscripcion_id": "n_inscripciones"})

p_8 = p_8[p_8["n_inscripciones"] >= 2]

p_8
```

### Punto 9. CASE WHEN / segmentación

Clasifica los cursos según su precio:
- `Bajo` si el precio es menor a 120
- `Medio` si el precio está entre 120 y 180 inclusive
- `Alto` si el precio es mayor a 180

Debes mostrar:
- `nombre_curso`
- `precio`
- `segmento_precio`

**Entregable:**
- SQL
- pandas

```python
# Punto 9 - SQL
query_9 = """
SELECT
    nombre_curso,
    precio,
    CASE
        WHEN precio < 120 THEN 'Bajo'
        WHEN precio BETWEEN 120 AND 180 THEN 'Medio'
        ELSE 'Alto'
    END AS segmento_precio
FROM cursos
"""

sql_9 = run_sql(query_9)
sql_9
```

```python
# Punto 9 - pandas
bins = [-float('inf'), 119, 180, float('inf')]
labels = ['Bajo', 'Medio', 'Alto']

cursos_segmentados = cursos.copy()
cursos_segmentados['segmento_precio'] = pd.cut(cursos['precio'], bins=bins, labels=labels)

p_9 = cursos_segmentados[['nombre_curso', 'precio', 'segmento_precio']]
p_9
```

### Punto 10. Ejercicio final integrador

Calcula el total pagado por categoría de curso.

Debes:
- unir `cursos`, `inscripciones` y `pagos`
- agrupar por `categoria`
- sumar `monto`
- ordenar de mayor a menor

**Entregable:**
- SQL
- pandas

```python
# Punto 10 - SQL
query_10 = """
SELECT
    c.categoria,
    SUM(p.monto) AS total_pagado
FROM cursos c
INNER JOIN inscripciones i ON c.curso_id = i.curso_id
INNER JOIN pagos p ON i.inscripcion_id = p.inscripcion_id
GROUP BY c.categoria
ORDER BY total_pagado DESC
"""

sql_10 = run_sql(query_10)
sql_10
```

```python
# Punto 10 - pandas
p_10 = cursos.merge(inscripciones, on="curso_id") \
             .merge(pagos, on="inscripcion_id")

resultado_final = p_10.groupby("categoria")["monto"] \
    .sum() \
    .reset_index() \
    .sort_values("monto", ascending=False) \
    .rename(columns={"monto": "total_pagado"})

resultado_final
```
