## Parte 2 — Selección y filtrado

**P4.** Selecciona solo las columnas `sepal_length` y `species` e imprime las primeras 5 filas.

```python
# Tu respuesta aquí
df[['sepal_length', 'species']].head()
```

**P5.** Filtra el DataFrame para quedarte solo con las flores de la especie `virginica`.  
¿Cuántas filas tiene ese subconjunto?

```python
# Tu respuesta aquí
df_virginica = df[df['species'] == 'virginica']

cantidad_filas = len(df_virginica)

print(f"El subconjunto de especie virginica tiene {cantidad_filas} filas.")
```

\*El subconjunto de especies virginica tiene 50 filas*

**P6.** Filtra las filas donde `petal_length` sea mayor a 5.0. Imprime cuántas filas cumplen esa condición.

```python
# Tu respuesta aquí
df_petal_may_5 = df[df['petal_length'] > 5.0]

cant_filas_grandes = len(df_petal_may_5)

print(f"Hay {cant_filas_grandes} flores con pétalos mayores a 5.0.")
```

*Hay 42 flores con pétalos mayores a 5*
