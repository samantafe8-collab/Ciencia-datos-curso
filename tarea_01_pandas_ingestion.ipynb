**¿Cuántas personas tienen la palabra Chief en el título de su trabajo?** Apoyese en los métodos para manipular strings y en el método `.apply()`

**Bono: ¿Existe una correlación entre la longitud del nombre del Título del trabajo y el Salario?**

crea una nueva columna donde se guarde la longitud del titulo del trabajo. tip: usar el método `.apply()`

```python
# Copie el código aca
```

```python
# Creo la nueva columna con la longitud del título del empleo, uso str(x) para por si hay valores Nulos
df['JobTitleLength'] = df['JobTitle'].apply(lambda x: len(str(x)))

# Visualizamos las primeras filas para verificar el resultado
print(df[['JobTitle', 'JobTitleLength']].head())
```

Utilice el método `.corr()` de Pandas para calcular la correlacion

```python
# Copie el código aca
```

```python
# Calculo la correlación entre las 2 columnas: JobTitleLength y TotalPayBenefits
corr_JTL_TPB= df[['JobTitleLength', 'TotalPayBenefits']].corr()

print(corr_JTL_TPB)
```

*En la matriz de correlación observamos valores menores a cero, lo que nos indica que no existe correlación entre las variables (columnas) JobTitleLength y TotalPayBenefits, es decir, la longitud del título no incide en el salario o pago total incluido beneficios.*
